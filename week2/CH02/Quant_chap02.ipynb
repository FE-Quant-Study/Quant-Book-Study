##1. 시장 데이터는 시장 환경을 반영한다. 
시장 데이터는 주문, 매매체결, 가격형성을 통제하는 방법, 규제를 포함하는 트레이팅 장소의 제도적 환경을 반영

- 알고리즘 트레이더는 ML을 포함한 알고리즘을 사용해 매수와 매도 주문의 흐름과 그 결과로 형성되는 거래량과 가격의 통계량을 분석해 트레이드 신호를 추출하며, 이를 이용해 수요/공급 동태 또는 특정 시장 참여자의 행태에 대한 이해를 높인다. 

### 1) 시장 미시 구조
시장 미시 구조는 제도적 환경이 트레이딩 과정에 어떻게 영향을 주고, 가격 발견, 매수-매도 호가, 일중 트레이딩 행태과 거래비용같은 결과를 어떻게 형성하는지 연구한다. 

### 2) 트레이딩 방법: 여러 종류의 주문
- 주문의 종류 : 시장가 주문, 지정가 주문, 스탑주문
- 주문의 조건: 모두 아니면 취소, 모두 체결 아니면 취소, 즉세 치결 아니면 취소, 재량 주문, 시가 또는 종가 시장 주문 

### 3) 트레이딩 장소: 거래소에서 다크풀까지
- 거래소: 매수자와 매도자가 각각 가장 낮은 매도 호가와 가장 높은 매수 호가를 위해 경쟁하는 중앙화된 시장
  - 쌍방 매매 또는 주문 주도 시스템에 의존
  - 대부분 특정 증권에 대한 시장을 조성하여 유동성을 제공하는 중개기관을 사용

- 오늘날 트레이딩은 파편화돼 있음. 미국에서는 2개의 주요 베뉴 대신 거래소와 전자통신망같은 대체거래시스템(ATS)을 포함한 13개 이상의 트레이딩 베뉴가 존재
  - ATS는 다크풀이 포함
  - 다크풀은 거래에 대한 정보를 거래 발생 후에 금융산업규제청에 보고

##2. 고빈도 데이터로 작업
통합피드는 각 트레이딩 베뉴로부터의 거래와 호가 데이터를 결합하며, 각 개별 거래소는 그 특정 베뉴에 대한 추가 활동 정보를 가진 독점적 상품을 공급한다,

###1) 나스닥 주문 호가창으로 작업
시장 데이터의 가장 중요한 소스는 호가창이며, 이는 모든 트레이딩 활동을 반영해 하루 종일 실시간 업데이트된다. 

- 미국주식시장의 호가 층
  - L1: 수많은 온라인 소스로부터 가능한 실시간 매수와 매도 호가 정보
  - L2: 특정 시장 조성자에 의한 매수 및 매도 호가와 함께 주어진 주식의 유동성에 대한 더 나은 통찰을 위해 최근 거래의 크가와 시간 정보를 더함
  - L3: 호가를 넣거나 변경할 수 있고 주문을 실행하고 거래를 확일할 수 있는 능력은 더함, 이는 시장 조성자와 거래소 회원 기업만 접근 가능함. L3 호가에 대한 접근성은 등록된 브로커가 최적 실행 요건을 축종하게함

##2) FIX 프로토콜을 활용한 거래의 통신
- FIX 프로토콜은 거래소, 은행, 브로커, 정산소와 기타 시장 참여자 간의 거래 체결 이전과 거래 체결 중 소통을 소통을 위한 사실상 메시징 표준

- 처음 글로벌 주식시장에서 널리 보급됐고, 외환, 채권, 파생시장으로 확대됐으며 더 나아가 일괄처리(STP)를 지원하는 사후 거래 시스템으로 발전

- 거래소는 FIX 메시지의 접근을 제공해, 알고리즘 트레이더들이 실시간 데이터 피드를 파스해 시장 행태를 추적 가능하게 하고, 사장 참여자의 현재 상황을 파악하고 다음 행태를 예상할 수 있게 함

- 메시지 시퀀스는 호가창을 재구성할 수 있게 만듬. 여러 거래소간 수많은 거래는 대규모 비정형 데이터를 생성 -> 이의 처리능력은 경쟁 우위성의 원천이 될 수 있음

- 프로토콜은 파이프 분리된 키-값의 쌍과 태그 기반의 FIXML구문을 지원

###3) 알파 타입의 필드는 format_alpha함수에 정의된 후 처리를 요함

```python
def format_alpha(mtype, data):
  for col in aplha_formats.get(mtype).keys():
    if mtype != 'R' and col == 'stock':
      data = data.drop(col, axis=1)
      continue
    data.loc[:,col] = data.loc[:, col].str.decode("utf-8").str.strip()
    if encoding.get(col):
      data.loc[:,col] = data.loc[:, col].map(encoding.get(col))
  return data
  
```
