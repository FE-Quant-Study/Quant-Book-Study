# Chapter 2. 시장 데이터와 기본 데이터: 소스와 기법

### 고빈도매매(초단타매매)
- 컴퓨터를 통해 빠른 속도로 내는 주문을 추선 번 반복하는 거래로서 알고리즘 매매 방식 중 하나
- 미리 정해놓은 특정한 조건을 충족하면 고성능 컴퓨터에 의해 빠른 속도로 주문이 자동으로 이루어짐
- 종류
    1. 시장조성(Market Making)<br>
        빠른속도로 매도호가와 매수호가를 내면서 주문을 내 호가 사이의 스프레드를 차지하는 전략
    2. 통계적 차익거래(Statistic Arbigrage)<br>
        주식과 주식, 주식과 선물 사이에 일시적으로 괴리가 발생할 때 빠른 매매를 통해 그 차익을 얻는 방법
    3. 이벤트 차익거래(Event Arbitrage)<br>
        신주인수권부사채 등이 발행될 대 주가에 일시적으로 발생하는 괴리에서 차익을 얻는 방법

### 데이터를 사용하는 전통전인 전략
- 비공개 데이터 세트를 구축하는데 초점이 아닌 공개된 데이터를 활용
- 기본 데이터에 초점을 두고, 기업에서 공시한 재무 데이터를 기반으로 재무 모델 구축
- 기술적 분석을 이용해 주가와 거래량으로부터 계산된 지표들을 사용해 시장 데이터에서 신호를 추출

> 머신러닝은 인간이 정의한 규칙이나 경험보다 더욱 효율적으로 시장 데이터와 기본 
데이터를 활용할 수 있음


### 시장 데이터는 시장 환경을 반영한다
- 시장 데이터  
    - 금융 수단에 의해 직접 또는 수많은 시장 중 하나의 시장에 있는 중개 기관을 통해 주문을 내는 방법, 처리되는 방법, 수요와 공급의 매칭에 의해 가격이 설정되는 방법의 산물
    - 주문, 매매 체결, 가격 형성을 통제하는 방법, 규제를 포함하는 트레이딩 장소의 제도적 환경을 반영

> 시장 데이터와 ML을 포함한 알고리즘을 사용해 매수와 매도 주문의 흐름과 그 결과로 형성되는 거래량과 가격의 통계량을 분석해 트레이드 신호 추출  
> ex(수요 공급 동태 / 특정 시장 참여자의 행태에 대한 이해)
<br>
- 시장 미시 구조: 기본 지식
    - 제도적 환경이 트레이딩 과정에 어떻게 영향을 주고, 가격 발견, 매수-매도 호가 스프레드와 호가, 일종 트레이딩 행태와 거래 비용 같은 결과를 어떻게 형성하는지 연구  
<br>
- 트레이딩 방법: 여러 종류의 주문
    - 시장가 주문 : 주문이 거래 플랫폼에 도달하는 순간 가능한 시장 가격으로 즉시 체결
    - 지정가 주문 : 시장 가격이 매도 지정가 주문의 경우 지정가보다 큰 경우에만 체결 / 시장 가격이 매수 지정가 주문의 경우 지정가보다 작은 경우에만 체결
    - 스탑 주문 : 매도 스탑의 경우 시장 가격이 지정한 가격보다 아래로 떨어지는 경우에만 활성화 / 매수 스탑 주문의 경우 시장 가격이 지정한 가격보다 위로 올라가는 경우에만 활성화
    - 모두 or 취소 주문 : 부분 체결을 방지할 수 있음. 특정 주식수를 한번에 체결 가능할 때에만 체결, 아니면 취소됨
    - 재량 주문 : 브로커가 체결의 시점과 가격을 결정
    - 시가 또는 종가 주문 : 시장의 개장 시점이나, 종료 시점 또는 그 근처에서 체결 (부분 체결 허용))


### 트레이딩 장소 : 거소에서 다크풀까지

- 증권 : 조직화되고 규제된 거래서 또는 다양한 정도의 형식을 가진 장외 시장에서 거래
- 거래소 : 매수자와 매도자가 각각 낮은 매도 호가와 높은 매수 호가를 위해 경쟁하는 중앙화된 시장

### 고빈도 데이터로 작업

- 나스닥 주문 호가창으로 작업
    - 시장 데이터의 가장 중요한 소스 => 호가창(실시간 업데이트)
    - 레벨 1(L1)  
        - 수많은 온라인 소스로부터 가능한 실시간 매수와 매도 호가 정보
    - 레벨 2(L2)  
        - 특정 시간 조성자에 의한 매수 및 매도 호가와 함께 주어진 주식의 유동성에 대한 더 나은 통찰을 위해 최근 거래의 크기와 시간에 대한 정보를 더함  
    - 레벨 3(L3)  
        - 호가를 넣거나 변경할 수 있고 주문을 실행하고 거래를 확인할 수 있는 능력을 더함
        - 시장 조성자와 거래소 회원 기업만 접근
        - 등록된 브로커가 최적 실행 요건을 충족하게 함

- FIX 프로토콜을 활용한 거래의 통신
    - 거래소, 은행, 브로커, 정산소와 기타 시장 참여자 간의 거래 체결 이전과 거래 체결 중 소통을 위한 메시징 표준
    - 거래소는 FIX 메시지의 접근을 제공하고, 알고리즘 트레이더들은 실시간 데이터 피드를 파싱해 시장 행태를 추적할 수 있음 ( => 메시지 시퀀스를 통해 호가창을 재구성할 수 있음)
    
- 나스낙 토탈뷰-잇치 데이터 피드
    - FIX 프로토콜이 압도적이지만 거래소 역시 자신들만의 프토로콜을 제공함
    - 나스닥 가입자들의 주식 주문부터 체결 또는 취소까지의 개별 주문을 추적할 수 있게 하는 토탈뷰-잇치(TotalView-ITCH) 를 통해 직접 데이터 피드 프토로콜을 제공함
    - 매수 또는 매도 호가인 호가창을 통해 얻을 수 있는 시장의 깊이(Market Depth)를 통해 유동성과 대규모 시장자 주문의 잠재적 가격 영향(Potential Market Impact)를 파악할 수 있음
