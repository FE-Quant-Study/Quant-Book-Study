##1. 투자업계에서 머신러닝의 부상
###1) 알고리즘 트레이딩과 머신러닝을 추진시킨 트렌드
- 시장 미시 구조의 변화: 전자거래확산, 자산군과 지역에 걸친 시장통합
- 자산군과 반대되는 리스크 팩터노출 측면에서의 투자전략 개발
- 컴퓨팅 파워, 데이터 생성, 관리, 통계적 방법의 혁명
- 인간의 재량적 투자와 비교해 알고리즘 트레이딩에서 선구자들의 성과

###2) 전자 거래에서 고빈도 매매까지(HFT)
- ECN: 주식 및 통화에 대해 브로커-딜로로 등록돼 지정된 가격으로 매수와 매도 주문을 매칭시키는 자동화된 대체거래시스템(ATS)

- 다크풀: 기관 투자자가 공개하지않고 대량주문을 거래할 수 있게 하는 사적 ATS


- 전자 거래의 부상과 함께 비용 효율적인 주문 집행 알고리즘이 빠르게 개발, 확산

- 고빈도 매매 : 마이크로초 범위의 매우 낮은 대기 시간과 매우 짧은 기간 동안 포지션을 유지하는 금융 상품의 자동화된 거래 
  - 거래 현장의 제도적 인프라와 시장 미시 구조의 비효율성을 탐지하여 활용
  - 패시브 전략이나 액티브 전략을 사용하여 거래당 작은 수익을 얻고자함

##3) 팩터 투자와 스마트 베타 펀드
- 자산이 제공하는 수익률은 금융투자와 연관된 불확실성이나 리스크의 함수
  - 주식투자 -> 기업의 사업리스크
  - 채권투자 -> 부도 리스크
  - 고유의 리스크 특성이 수익률을 예측할 수 있는 범위 내에서 리스크 팩터의 형태를 식별하고 예측해 투자 전략을 설계할 때 중요한 초점이 됨

- 현대 포트폴리오 이론(MPT)은 특정 자산에 대한 고유한 리스크와 리스크 원천을 구별해냄
  - 고유 리스크는 분산화로 제거가능
  - 체계적 리스크는 제거 불가능
  -> 자산은 기초 공통 리스크에 노출되기 때문에 리스크 프리미엄을 가짐

##4) 알고리즘 개척자는 인간보다 우위에 있다. 
- 투자 산업이 임의일임 스타일에서 퀀트 스타일로 전환
  - 디지털 데이터의 가용성에 있어 기하급수적 증가
  - 낮은 비용으로 컴퓨팅 성능과 데이터 저장용량 증가
  - 복잡한 데이터 세트를 분석하기 위한 통계적 방법의 발전

##5) 머신러닝과 대체 데이터
- 머신러닝을 이용한 정보 우위는 다양한 데이터 원천을 대량으로 수집하고 실시간으로 분석하는 능력과 관련

- 대체 데이터: 거래 신호를 포함하는 모든 데이터 소스를 포함
  - 일상적인 업무 과정에서 생성된 데이터를 가치있는 지능 데이터로 전환하는 것뿐만 아니라 위성이미지, 신용카드 판매, 감성 분석, 모바일 위치 정보데이터, 웹사이트 스크래핑과 같은 소스 포함

#2. 머신러닝 기반 전략의 설계와 실행
- 머신러닝 사용 사례에 적용되는 중요한 기술과 기법
  - 중요한 데이터 소스를 소싱하고 관리하는 방법
  - 신호 내용을 추출하는 정보 특성이나 알파 팩터를 엔지니어링하는 방법
  - 포트폴리오를 관리하고 전략 성과를 추적하는 방법
image.png

###1) 데이터의 소싱과 관리
- 알파 신호를 포함해 시장, 펀드멘탈, 대체 데이터의 소스를 식별, 평가
- 클라우드 기반의 확장가능한 데이터 인파라와 하둡, 스파크와 같은 분석 도구를 배포하거나 액세스해 빠르고 유연한 데이터 액세스를 용이하게 함
- PLT기준으로 원하는 빈도로 조정하여 선견자 편향을 피하기 위해 데이터를 조심스럽게 관리하고 식별
  - 데이터는 주어진 시간에 사용할 수 있고 알려진 정보만반영해야함

###2) 알파 팩터 리서치에서 포트폴리오 관리까지
- 알파팩터는 데이터에서 신호를 추출해 거래 기간동안 주어진 투자 영역에 대한 수익을 예측하도록 설계됨
image.png

- 리서치 스텝
  - 리서치 스텝은 알파 팩터의 설계와 평가가 포함됨
  - 예측 팩터는 데이터 소스와 자산 수익률 등 중요한 전략 입력 변수 간의 체계적인 관계의 일부 측면을 포착함
  - 예측력 최적화를 위해서는 창의적 특성공학 필요
  - 데이터 마이닝으로 인한 잘못된 발견은 신중한 관리가 필요한 주요 리스크

- 실행 스템
  - 알파팩터는 매수 주문이나 매도 주문으로 이어지는 신호를 발산 
  - 포트폴리오 보유는 총 포트폴리오 리스크에 상호작용하고 기여하는 특정 리스크 프로파일을 가짐

###3) 전략 백테스팅
- 벡테스팅 엔진: 편향을 갖지 않는 성과 추정치를 얻고자 현실적인 실행을 시뮬레이션
  - 시장 상황에 따라 거래-신호 평가, 주문집행, 실행을 실무적 관점에서 정확히 나타낼 필요성이 있

##3. 트레이딩을 위한 머실러닝
- 알고리즘 트레이딩의 진화
  - 1980 학술적 연수로부터 신호가 나옴
  - 2000 팩터 기반의 투자 확산
  - 머신러닝 능력과 대체 데이터에 대한 투자 ㅇ추진

- 트레이터의 알고리즘을 사용해 규칙을 실행할 때의 목표
  - 우호적 가격 결정을 목표로 하는 거래 실행 알고리즘
  - 작은 가격변동으로부터 이익을 얻는 단기 거래
  - 다른 시장 참여자의 행동을 예측하는 것을 목표로 하는 형태 거래 전략
  - 절대 및 상대 가격을 기반으로 하는 거래 전략과 수익률 예측

- 거래 실행 프로그램: 거래의 시장 영향 제한 목표
- 고빈도 매매펀드 : 매수-매도 스프레드 기반으로 작은 가격 변동, 통계적 차익 거래로부터의 이익을 얻고자 매우 짧은 보유 기간에 가장 많이 의존
- 행동주의 알고리즘: 상당한 가격영향을 미치는 더 큰 플레이어의 움직임을 예상


###1) 거래를 위한 머신러닝 사용 사례
- 패턴을 확인하고 특성을 추출해 통찰력을 생성하는 데이터 마이닝
- 리스크 팩터나 알파와 거래 아이디어를 창출하기 위한 지도학습
- 개별 신호를 전략으로 통합
- 알고리즘으로 학습된 리스크 프로파일에 따른 자산 배분
- 합성 데이터의 사용이 포함된 전략의 테스트와 평가
- 강화학습을 사용한 자동화된 대화형 전략 개선
